```python
fuel_price=110
friends=4
km_per_litre=15
distance=300
toll_cost=500
food_cost=1300
budget=1000

fuel_cost=(distance/km_per_litre)*fuel_price
tot_trip_cost=fuel_cost+toll_cost+food_cost
cost_per_person=tot_trip_cost/friends
lp=budget>=cost_per_person

print(f"Fuel Cost:{fuel_cost} Total Trip Cost= {tot_trip_cost} Cost per Person:{cost_per_person} ,{lp}")
```
