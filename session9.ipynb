```python
#zip function
quantities =[2,5,4,3,2]
prices= [41.25,65,85,2,3]

zipped_items =list(zip(quantities,prices))
print(list(zipped_items))

total_cost= list(map(lambda item : item[0]* item[1] ,zipped_items))
print(total_cost)
print(sum(total_cost))
```

```python
class BankAccount:
  accStatus= "Active"
  def __init__(self,owner,accno,bal):
    self.owner=owner
    self.accno =accno
    self.bal =bal

  def deposit(self,amount):
    if amount > 0:
      self.bal += amount
      print(f"Deposited {amount:.2f} Balance: {self.bal:.2f}")
    else:
      print("enter a valid amount")

  def withdraw(self,amount):
    if amount <= 0:
        print("Invalid Amount: Please enter a valid Amount.")
    elif self.bal >= amount:
        self.bal -= amount
        print(f"Withdrawn : {amount:.2f} Balance: {self.bal:.2f}")
    else:
        print("Insufficient Balance")

  def display(self):
    print(f"Owner :{self.owner} ACCno :{self.accno} Balance:{self.bal}")


cst1=BankAccount(101,"ACC101",43003)
cst2=BankAccount(102,"ACC102",86400)


cst1.deposit(123)
cst1.withdraw(-2)
cst1.display()

cst2.display()


```

```python
class SavingsAccount(BankAccount):
  def __init__(self,owner,accno,bal,interest_rate,min_balance):
    super().__init__(owner,accno,bal)
    self.interest_rate= interest_rate
    self. min_balance = min_balance

  def calc_interest(self):
    interest= self.bal * self.interest_rate
    self.bal += interest

  def withdraw(self,amount):
    if self.bal - amount >= self.min_balance:
      super().withdraw(amount)
    else:
      print("Insufficient Balance")


cst1 =SavingsAccount(101,"ACC101",43126,0.8,1000)
cst1.deposit(200)
cst1.calc_interest()
cst1.display()



```
