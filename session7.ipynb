```python
student={
    'name':'Thomas',
    'mark':{
        'Maths':85,
        'Science':98,
        'English':92,
        'History':100
    },
    'cgpa':8.8,
    'grad_year':2026

}
```

```python
avg_score=sum(student['mark'].values())/len(student['mark'])
avg_score
```

```python
if avg_score >= 90:
  grade="A+"
  if student['cgpa'] >= 9:
    status="Exellent"
  else:
    status="Very Good"
elif avg_score >= 75:
  grade="A"
  if student['cgpa'] >=7:
    status="Good"
  else:
    status="Okay"
elif avg_score >= 60:
  grade="B"
  if student['cgpa'] >=5:
    status="Satisy"
  else:
    status="Failed"

print(status,grade)

```

```python
name=student['name']
if avg_score >= 90:
  grade="A+"
  if student['cgpa'] >= 9:
    status="Exellent"
  else:
    status="Very Good"
elif avg_score >= 75:
  grade="A"
  if student['cgpa'] >=7:
    status="Good"
  else:
    status="Okay"
elif avg_score >= 60:
  grade="B"
  if student['cgpa'] >=5:
    status="Satisy"
  else:
    status="Failed"

if status!="Failed":
  gdstat="Graduated"
else:
  gdstat="Ineligible"

print(f"Name: {name}")
print(f"Grade: {grade}")
print(f"Marks: {student['mark']}")
print(f"Total Marks:{sum(student['mark'].values())}")
print(f"Average Score :{sum(student['mark'].values())/len(student['mark']) }")
print(f"Graduation:{gdstat}")

```

```python
p1=int(input("enter the price for p1 "))
p2=int(input("enter the price for p2 "))
p3=int(input("enter the price for p3 "))
p4=int(input("enter the price for p4 "))
p5=int(input("enter the price for p5 "))

total=p1+p2+p3+p4+p5

if total > 2000:
  final_price = total * (1 - 20 / 100)
elif total > 21000:
  final_price = total * (1 - 10 / 100)
else:
  print("No Discount")
print(f"Final Price:{final_price}")


```
