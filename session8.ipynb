```python
def add(a,b):
  sum=a+b
  return sum

def sub(a,b):
  d=a-b
  return d

def mul(a,b):
  m=a*b
  return m

def div(a,b):
  div=a/b
  return div


a= int(input("enter the 1st number"))
b= int(input("enter the 2nd number"))

add(a,b),sub(a,b),mul(a,b),div(a,b)
```
